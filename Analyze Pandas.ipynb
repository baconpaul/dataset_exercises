One of the schools has "pizza thursday", a day when 5 cents for each pizza slice sold goes
to improve the computer lab and music studio at the school. On this day, students prefer
pizza more than thay do in other weeks. Which school does Pizza Thursday?

```python
foodOnly = dset.drop( [ "Enrollment", "Attendance" ],axis=1)

dailyTrend = foodOnly.groupby( ["School","DayOfWeek"] ).agg( "mean" )
dailyTrend
```

Now for each school we will make the weekly average excluding thursdays and for thursdays only 
and then figure out which school has it's thursday different by it's other days by the most.

```python
# This is an unofrunate bit of syntax
fl = dailyTrend.reset_index()

# So we will make the thursday and no thursday sets
noThu = fl[ fl["DayOfWeek" ] != "04-Thu" ]
thu = fl[ fl[ "DayOfWeek"] == "04-Thu" ]

# Then find the average by school
noThuBySchool = noThu.groupby( "School" ).agg( "mean" )
thuBySchool = thu.groupby( "School" ).agg("mean") 

# Then take a look at the thursday pizza vs non-thursday pizza averages proportional
# to the average sales
thursdayDifferences = ( thuBySchool.Pizza - noThuBySchool.Pizza ) / noThuBySchool.Pizza
thursdayDifferences
```

```python
# And the name of the school is (again, too bad we need that 'to-frame')
pizzaSchool = thursdayDifferences.to_frame().Pizza.idxmax()
pizzaSchool
```

```python
pizzaSchoolData = fl[ fl[ "School" ] == pizzaSchool ]
pizzaSchoolData.Pizza.plot()
pizzaSchoolData.Hamburger.plot()
pizzaSchoolData.HotDog.plot()
pizzaSchoolData.Skipped.plot()
```
